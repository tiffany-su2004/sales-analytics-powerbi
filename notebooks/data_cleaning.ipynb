**â­ Step 4.1 â€” Load the raw data (with correct encoding)**

```python
import pandas as pd

# Load with proper encoding
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

# Always check the first few rows
df.head()
```

**â­ Step 4.2 â€” Convert ORDERDATE into a real datetime**

Why:
Power BI cannot perform:

time-series

YoY

MoM

forecasting
â€¦unless the date column is a proper datetime.

Your dataset has ORDERDATE as a string (object), e.g. "2/24/2003 0:00".

```python
# Convert ORDERDATE from string â†’ datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

**â­ Step 4.3 â€” Standardize column names**

Why:
BI systems prefer clean column names:

no spaces

no uppercase

consistent pattern

Example:
"ORDERNUMBER" â†’ "ordernumber"
"PRICE EACH" â†’ "priceeach"

```python
df.columns = df.columns.str.lower().str.strip()
```

**â­ Step 4.4 â€” Handle Missing Values (Strategic Approach)**

**â­ Handle Missing Values (Strategic Approach)**
Let's analyze your missing values:
addressline2 â†’ 2521 missing  (not useful)
state â†’ 1486 missing          (optional)
territory â†’ 1074 missing      (optional)
postalcode â†’ 76 missing        (can be left null)

âœ” Our decision:

Keep missing values as NaN because:

They do NOT affect KPIs

They do NOT affect fact table

Power BI can still group by COUNTRY and CITY

These fields arenâ€™t used in any formula

```python
# Drop columns that provide zero analytical value
df = df.drop(columns=["addressline2"])
```

**â­ Step 4.5 â€” Create Profit Metrics (Optional but very powerful)**

Your dataset has:

priceeach

msrp

sales

We can create:

profit = sales â€“ (cost)

We donâ€™t know cost directly, but MSRP (full retail price) can approximate margin.

So:
profit = (msrp - priceeach) * quantityordered

```python
df['profit'] = (df['msrp'] - df['priceeach']) * df['quantityordered']
```

**â­ Step 4.6 â€” Create Date Columns (Year, Month, Quarter)**

Even though you already have YEAR_ID, MONTH_ID, QTR_IDâ€¦
we prefer using the real date column because:

YEAR_ID may not match actual date

ORDERDATE enables more accurate comparisons

Power BI time intelligence functions require a date table

Explanation:
This creates cleaner dimensions for building dim_date.

```python
df['year'] = df['orderdate'].dt.year
df['month'] = df['orderdate'].dt.month
df['quarter'] = df['orderdate'].dt.quarter
df['month_name'] = df['orderdate'].dt.strftime('%B')
```

**â­ Step 4.7 â€” Standardize categorical text columns**

Columns like COUNTRY, STATE, PRODUCTLINE, STATUS should be cleaned for:

capitalization consistency

trailing spaces

formatting issues

Example:

"usa" â†’ "Usa"

" France " â†’ "France"

Why:
Clean categories = clean grouping in Power BI visuals.


```python
cat_cols = ['status','productline','customername','city','state','country','territory','dealsize']

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

**â­ Step 4.8 â€” Remove obvious duplicates**

```python
df = df.drop_duplicates()
```

**â­ Step 4.9 â€” Sort the data by date**

```python
df = df.sort_values(by='orderdate')
```

**â­ Step 4.10 â€” Export the cleaned dataset**

```python
df.to_csv("cleaned_sales.csv", index=False)
```
